## Analysis
#### I found several interesting trends as far as latitudes effect on the weather on each side of the equator
- As is expected from the equator to 60 degrees north latitude temperature decreases, however the drop off isnt as significant in the southern hemisphere, this is possibly because there are less cities closer to the south pole to see that temperature drop or the northern hemisphere is just cooler.
- Humidity in the southern latitudes is mostly always above 40% where as northern latitudes above 20 degrees north have much lower percentages.
- Cloudiness is extremely uniform across all latitudes clustering on 0% and 100% cloud cover this is expected as clouds move all across the planet and can form anywhere theres water.

```python
!pip install citipy
#dependencies
import pandas as pd
import openweathermapy.core as own
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
from pprint import pprint
import requests
import random
import pandas as pd
from config import api_key
from citipy import citipy
#api key for open weather
from config import api_key
```

```python
cities = citipy.WORLD_CITIES_DICT

```

```python
#creating arrays to store information for the dataframe
units="imperial"
city = []
name=[]
code=[]
latitude=[]
longitude=[]
humidity=[]
clouds=[]
temp=[]
wind=[]
coordlist= list(cities.keys())
#looping through a list of 500 random cities from the citipy dictionary of world cities
for i in range(500):
    randomcity = random.choice(coordlist)
    city.append(randomcity)
    citylist=citipy.nearest_city(*city[i])
    name.append(citylist.city_name)
    code.append(citylist.country_code)
    lat,lon= city[i]
    latitude.append(lat)
    longitude.append(lon)
    #making an api call for each lat and lon I got from citipy then storing values in the arrays
    url= "http://api.openweathermap.org/data/2.5/weather?"
    queryurl=f"{url}lat={latitude[i]}&lon={longitude[i]}&units={units}&appid={api_key}"
    response = requests.get(queryurl)
    json=response.json()
    temp.append(json["main"]["temp_max"])
    humidity.append(json["main"]["humidity"])
    wind.append(json["wind"]["speed"])
    clouds.append(json["clouds"]["all"])
    print(f"{name[i]}.")
    print(f"{code[i]}.")
    print(queryurl)
```

```python
#Creating the dataframe from the arrays
Weatherdf=pd.DataFrame({"City":name,
                       "Country Code":code,
                       "Latitude":latitude,
                       "Longitude":longitude,
                       "Temperature (F)":temp,
                       "Humidity":humidity,
                       "Cloudiness":clouds,
                       "Wind Speed (MPH)":wind},
                      columns=["City","Country Code","Latitude","Longitude","Temperature (F)","Humidity","Cloudiness","Wind Speed (MPH)"])
Weatherdf.to_csv("weatheroutput.csv",sep=',',encoding="utf-8")
Weatherdf.head()
```

```python
#Charting Lat vs the info for the next 4 steps
plt.scatter(Weatherdf["Latitude"],Weatherdf["Temperature (F)"],marker="o", facecolor= "red", edgecolors="black",alpha=1,)
plt.title("Latitude vs Temp (F)")
plt.xlabel("Latitude")
plt.ylabel("High Temperature")
plt.savefig("Latitude vs Temp.png")
plt.show()
```
